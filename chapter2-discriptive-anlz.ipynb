```python
genre_release_time_count = df.groupby(['Genre', 'ReleaseTime']).size().reset_index(name='Count')
genre_release_time_count
```

```python
df['ReleaseDate']=pd.to_datetime(df['ReleaseDate'],format='%d-%b-%y')
df['Month'] = df['ReleaseDate'].dt.month_name()
month_counts = df['Month'].value_counts()
max_release_month = month_counts.idxmax()
max_release_count = month_counts.max()

print("Month with maximum movie releases:", max_release_month)
print("Number of movies released in that month:", max_release_count)

 
```

```python
df[df['Budget']>25][['MovieName','Month','Budget']]
```

```python
df['Roi']=df['BoxOfficeCollection']-df['Budget']
df.head(5)
```

```python
df[['MovieName','Roi']].sort_values('Roi',ascending=False)[0:10]
```

```python
roi_by_release_time = df.groupby('ReleaseTime')['Roi'].mean().reset_index()
roi_by_release_time = roi_by_release_time.sort_values('Roi', ascending=False)
print(roi_by_release_time)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(1, 2, 1)
plt.hist(df['Budget'], bins=10, edgecolor='black', color='skyblue')
plt.title('Histogram of Movie Budgets')
plt.xlabel('Budget (in Crores)')
plt.ylabel('Frequency')
```

```python
plt.subplot(1, 2, 2)
sns.kdeplot(df['Budget'], shade=True, color='orange')
plt.title('Distribution Plot of Movie Budgets')
plt.xlabel('Budget (in Crores)')
```

```python
df['Genre'].unique()
```

```python
filtered_df = df[df['Genre'].isin(['comedy', 'drama'])]
print(filtered_df['Genre'].value_counts())
plt.subplot(1, 2, 1)
sns.boxplot(x='Genre', y='Roi', data=filtered_df, palette='pastel')
plt.title('Box Plot of ROI for Comedy vs Drama')
plt.xlabel('Genre')
plt.ylabel('ROI')
```

```python
plt.subplot(1, 2, 2)
sns.kdeplot(data=filtered_df[filtered_df['Genre'] == 'comedy']['Roi'], label='Comedy', shade=True, color='orange')
sns.kdeplot(data=filtered_df[filtered_df['Genre'] == 'drama']['Roi'], label='Drama', shade=True, color='blue')
plt.title('Distribution Plot of ROI for Comedy vs Drama')
plt.xlabel('ROI')
plt.legend()
```
